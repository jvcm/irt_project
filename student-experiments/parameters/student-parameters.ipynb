```python
import os
import pandas as pd
import numpy as np
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import glob
from scipy.stats import spearmanr

sns.set_context('paper')
cmap1 = sns.cubehelix_palette(rot=-.5,light=1.5,dark=-.5,as_cmap=True)
```

```python
filenames = os.listdir('./')
df_abilities = pd.concat([pd.read_csv(filename, index_col=0) for filename in filenames if 'abilities.csv' in filename]).reset_index(drop=True)
df_items = pd.concat([pd.read_csv(filename, index_col=0) for filename in filenames if 'parameters.csv' in filename]).reset_index(drop=True)
```

```python
fig = plt.figure(figsize=(5,5))
c = df_items['discrimination']
mean_error = df_items['mean_error']
diff = df_items['difficulty']
plt.scatter(diff, mean_error, c=c, cmap=cmap1, edgecolor='k', s=30)
# plt.text(x = 0, y = 0, s='Mean = %.2f Std = %.2f Corr = %.2f' %(mean_error.mean(), mean_error.std(), spearmanr(a = diff, b = mean_error)[0]), fontsize=10)
plt.xlabel('Difficulty')
plt.ylabel('Mean Error (Item)')
plt.xlim([0,1])
plt.ylim([0, 3])
fig.savefig('./diff_error')
```

```python
fig = plt.figure(figsize=(5,5))
c = df_abilities['grade']
mean_error = df_abilities['mean_error']
ab = df_abilities['ability']
plt.scatter(ab, mean_error, c=c, cmap=cmap1, edgecolor='k', s=30)
# plt.text(x = 0, y = 0, s='Mean = %.2f Std = %.2f Corr = %.2f' %(mean_error.mean(), mean_error.std(), spearmanr(a = ab, b = mean_error)[0]), fontsize=10)
plt.xlabel('Ability')
plt.ylabel('Mean Error (Student)')
plt.xlim([0,1])
plt.ylim([0, 3])
fig.savefig('./ab_error')
```
